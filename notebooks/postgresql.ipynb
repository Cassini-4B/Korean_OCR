```python
import pandas as pd
import numpy as np
from __future__ import print_function, division


import pickle

# Get pandas and postgres to work together
import psycopg2 as pg
import pandas.io.sql as pd_sql

# We are also going to do some basic viz
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Postgres info to connect

connection_args = {
    'host': '18.218.246.180', # You'll have to update this to your IP
    'user': 'ubuntu',    # username
    'dbname': 'korean_ocr',   # DB that we are connecting to
    'port': 5432         # port we opened on AWS
}

# We will talk about this magic Python trick!
connection = pg.connect(**connection_args)
```

```python
query = "SELECT * FROM information_schema.columns WHERE table_name = 'letters_merged'"


columns = pd_sql.read_sql(query, connection)
```
